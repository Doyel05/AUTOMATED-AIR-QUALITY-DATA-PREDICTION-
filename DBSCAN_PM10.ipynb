```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("PRSA_Data_Aotizhongxin_20130301-20170228.csv")
```

```python
df
```

```python
df = df.drop(['No','wd','station'], axis = 1)
```

```python
df
```

### 0th Hour PM10

```python
df0 = df[df.hour == 0]
```

```python
df0
```

```python
df0.shape
```

```python
df0.isnull().values.any()
```

```python
df0 = df0.replace(np.nan,0)
```

```python
df0
```

```python
df0.isnull().any()
```

```python
data = df0.iloc[:,[2,5]].values
```

```python
data.shape
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors = NearestNeighbors(n_neighbors = 5)
```

```python
nbrs = neighbors.fit(data)
```

```python
distances, indices = nbrs.kneighbors(data)
```

```python
distances = distances[:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
distances = np.sort(distances, axis = 0)
```

```python
plt.rcParams['figure.figsize'] = (10 , 6)
plt.plot(distances)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN (eps = 18, min_samples = 5).fit(data)
```

```python
labels_0 = dbscan.labels_
```

```python
plt.scatter(data[:, 0], data[:, 1], s=10, c = labels_0, cmap = 'viridis')
plt.title("Scatter Plot of PM 10 value on 0th hour over days from 2013 to 2017")
plt.xlabel("Days")
plt.ylabel("PM 10 Values 0th Hour")
plt.grid()
plt.show()
```

```python
np.unique(labels_0)
```

```python
num_clusters = len(set(labels_0))-(1 if -1 in labels_0 else 0)
num_noise = list(labels_0).count(-1)

print("Number of Clusters made : ", num_clusters)
print("Number of Noise present : ", num_noise)
```

```python
data1 = data[labels_0 == -1,0]
data2 = data[labels_0 == -1,1]
```

```python
plt.title("Noise values in PM 10")
plt.xlabel("Days")
plt.ylabel("PM 10 values on 0th Hour")
plt.grid()
plt.scatter(data1, data2, s=10, c="red")
```
